<h1 id='0-problem-description' style='color:#7159c1; border-bottom:3px solid #7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-weight: bold; text-align:left; font-size:240%;padding:0'>üìù | Problem Description</h1>

Recommendation Systems are present everyday in our lives helping us to find the perfect anime to watch on Crunchyroll, or to binge watch that show on Netflix, keep watching related YouTube videos or even to buy new products on Amazon. I can spend the whole day listing examples about this system, but I think that you got the glimpse where Recommendation Systems are used, don't you?

However, have you ever get bored on Netflix and Instagram because they only recommend items related to the same topic? It is not because I liked two videos about cars on Instagram that I am a huge fan of cars and want to see only reels and posts about it. Or it is not because I saw a gamer chair price on Amazon that I want to buy only gamer chairs. When you keep receiving recommendations only related to a especific topic or product, we say that you are into a Limited Bubble of Recommendations.

Limited Bubble of Recomendations tends to make companies lose customers, so, in order to minimize these bubbles and to expand the recommendations' diversity, algorithms have been tuned and new variables have been considered to the model.

Nowadays, there are four main Recommendation Algorithms: `Demographic Filtering, Content-Based Filtering, Collaborative Filtering and Hybrid Filtering`, let's see how each one works assuming that we are dealing with Animes Recommendation.

<p style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>- Demographic Filtering</p>

> **Logic** - `it recommends animes that are popular between all users. Normally recommends the animes with higher Bayesian Mean Score, Arithmetic Mean Score and Popularity Rank`;

> **Example** - `if Bleach, Oregairu and Jujutsu Kaisen are the animes with higher scores on a stream platform, this platform will recommend them to all users, regardless the taste of the users`;

> **Metrics** - `Arithmetic Mean Score (when dealing with few ratings data), Bayesian Mean Score (when dealing with many ratings data) and Popularity Rank (when dealing only with popularity and not ratings)`;

> **Advantages** - `only animes data is required, excluding user data; the model training takes less time to compute; the metric is not complicated to understand`;

> **Disadvantages** - `bad recommendations are made since this technique generalizes the taste of all users to only one: the animes with higher Score and Popularity`.

    <figcaption>Figure 1 - Demographic Filtering Diagram. By <a href='https://www.width.ai/post/recommender-systems-recommendation-systems'>Matt Payne - Recommender Systems For Business - A Gentle Introduction¬©</a>.</figcaption>

<br /><br />

<p style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>- Content-Based Filtering</p>

> **Logic** - `it recommends animes that are similar to other animes that a user liked. There are two variations of this technique: Plot Description Based (it recommends animes with similar synopsis) and Metadata Based (it recommends animes with similar metadata, such as genres, licensors, author, producers and studios)`;

> **Example** - `if a user liked Dragon Ball Z, there is a high probability that the platform will recommend Dragon Ball Super or Naruto (both shounen animes with superpowers)`;

> **Metrics** - `Cosine Similarity Score and Pearson Correlation. In this project, we will be working with Cosine Similarity Score`;

> **Advantages** - `better recommendations are provided compared to Demographic Filtering Technique; it recommends similar items that the users liked; each user receives personalized recommendations`;

> **Disadvantages** - `more data is required, such as synopsis and metadata; when working on Plot Description Based, there is a probability to get only sequels and prequels recommendations; when working on Metadata Based, there is a probability to create a Bubble that only recommends animes with an especific genre`.

</figure>

<br /><br />

<p style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>- Collaborative Filtering</p>

> **Logic** - `it recommends animes that are similar to animes liked by similar users. There are two variations of this technique: User Based Filtering (when dedaling with users similarity) and Item Based Filtering (when dealing with items similarity)`;

> **Example** - `if user A and user B liked Demon Slayer, user A is considered similar to user B and user B also like Grand Blue, the platform will recommend Grand Blue to user A`;

> **Metrics** - `Cosine Similarity Score and Pearson Correlation. In this project, we will be working with Cosine Similarity Score`;

> **Advantages** - `better recommendations are provided compared to Content-Based and Demographic Filtering; it recommends similar items that similar users liked; each user receives personalized recommendations; there is a few probability to create a Bubble of Contents`;

> **Disadvantages** - `anime and user data are required; the dataset must contain a significant amount of user ratings to animes; there is yet a probability of creating a Bubble of Contents; it is needed to add a cut-off amount of ratings and mean rating score by user in order to avoid outliers; the algorithm is more complex and takes more time to run`.

</figure>

<br /><br />

<p style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>- Hybrid Filtering</p>

> **Logic** - `it combines the Content-Based and Collaborative Filtering techniques. It normally applies the first one when there are few users ratings available and smoothly replaces it to the second one as more users ratings become available`;

> **Example** - `1) When few users ratings are availabe: if a user liked Dragon Ball Z, there is a high probability that the platform will recommend Dragon Ball Super or Naruto (both shounen animes with superpowers); 2) When many users ratings are available: if user A and user B liked Demon Slayer, user A is considered similar to user B and user B also like Grand Blue, the platform will recommend Grand Blue to user A`;

> **Metrics** - `Cosine Similarity Score and Pearson Correlation. In this project, we will be working with Cosine Similarity Score`;

> **Advantages** - `since it combines two techniques, it contains the advantages related to both of them; also, it decreases the probability of making poor recommendations when the platform is new and does not contain many users ratings`;

> **Disadvantages** - `since it combines two techniques, it contains the disadvantages related to both of them; also, it requires more computational resources since the algorithm is more complex and complete than the previous techniques`.

<h1 id='4-metrics' style='color:#7159c1; border-bottom:3px solid #7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-weight: bold; text-align:left; font-size:240%;padding:0'>üìè | Metrics</h1>

Two metrics were used to evaluate the Models: `Bayesian Mean` for Demographic Filtering, and `Cosine Similarity` for Content-Based, Collaborative and Hybrid Filtering.

<p style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>- Bayesian Mean</p>

When working with mean scores of products, we have a tendency to use the `Arithmetic Mean` rather than other metrics, however, it can be a problem for Recommendation System Models due to the difference between the score value and the amount of scores. For instance, picture an anime that has a Mean Score of 9.8 and only 100 votes, and another anime with a Mean Score of 8.5 and 100,000 votes. Which one is better? The one with the higher score and less votes or the one with the smaller score and more votes? In this case and statistically speaking, it is correct to assume that the one with the smaller score and more votes is the better because the difference of the amount of votes between the two animes is huge. So, we have just stumbled upon a problem: there are situations that the Arithmetic Mean is not suitable and can compromisse Recommendation Systems Models.

Having this in mind, we can use the `Cumulative Mean`. Its idea is similar to the Arithmetic one, but the difference is that the value is not divided by the number of votes. So, if an anime got two ratings of 10, we say that its Cumulative Mean is 20, while its Arithmetic Mean is 10. But, this method has a problem too: consider an anime with 1,000,000 ratings of 1 score and another anime with 10,000 ratings of 10 score, if we calculate their Cumulative Mean, the first anime will have a score of 1,000,000 whereas the second one will have a score of 100,000. So, even though having the better ratings, the anime can be in a bad rank position if the number of votes is substancial lower than an anime with worse ratings and a huge number of votes.

So, up to now, we have this two metrics problems:

> **Arithmetic Mean** - `does not assume that an anime with lower score and huge number of ratings is better than an anime with higher score and only a few number of ratings`;

> **Cumulative Mean** - `does not assume that an anime with a lower score and huge number of good ratings is better than an anime with higher score and a more huge number of bad ratings`.

<br /><br />

In order to solve these two problems and to giver better, realistic mean scores, the `Bayesian Mean` has been created!! Its idea is simple, first we have to find the number of ratings of the item, the Arithmetic Mean of the item, the minimum number of ratings that an item must have to be considered in the system and the Arithmetic Mean across all items. After that, we apply them in the following equation:

```python
bayeasian_mean = ((v / (v + m)) * A) + ((m / (v + m)) * C)
```

$$
Bayesian Mean = \frac{v}{v + m} \cdot A + \frac{m}{v + m} \cdot C
$$

where:

- v: number of the item's ratings;
- m: minimum number of ratings that an item must have to be considered in the system;
- A: item's Arithmetic Man;
- C: Arithmetic Mean across all items.

Due to its good results, Bayesian Mean is used by IMDB to calculate the mean score of movies listed on the website.

Also, it is highly recommended to start using Arithmetic Mean when there are a few ratings to an item and smoothly migrate to Bayesian Mean accordingly the number of ratings increase.

<p style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>- Cosine Similarity</p>

`Cosine Similarity` is a perfect metric to measure the similarity between two values of a same feature, especially when dealing with sequential texts or rating scores. In a nutshell, it works dividing the sum of the multiplication of the appearance of the elements in each items by the product of the squared root of the quadratic appearance of each element into the items ü§Øüò∂‚Äçüå´Ô∏è. You can be wondering: *What the hell did I just read?*. Do not worry, The example below makes everything clear ü§£

Consider these two sentences: `Hello World!` (sentence A) and `Hello!` (sentence B). The first thing to do is to create a table showing the frequency of each word in each sentence:

        <th style='border-style: solid'># in B</th>

    </tr>
</table>

After that, we: 

- plot the frequencies using points;
- trace two lines across the points, one line to one point; 
- calculate the angle between the lines;
- calculate the cosine of the angle;
- the cosine of the angle is the Cosine Similarity Score. It goes from 0 (completely not similar) to 1 (completely similar).

In this example, the angle between the lines is `45¬∫` and its cosine is `0.71`, thus we can tell that both sentences are 71% similar. The image below ilustrates the plot and the calculation:

<br /><br />

Now that we already know how it works, let's take a peek at its equation:

```python
sum(A[i] * B[i]) / (sqrt(sum(A[i]**2)) * sqrt(sum(B[i]**2)))
```

$$
Cosine Similarity = \frac{\sum_{i=1}^{n} (A[i] \cdot B[i])} {\sqrt{\sum_{i=1}^{n} (A[i]^2)} \cdot \sqrt{\sum_{i=1}^{n} (B[i]^2)}}
$$

where:

- A and B: items;
- n: number of different elements into the items;
- A[i]: number of times the element "i" is present into item A;
- B[i]: number of times the element "i" is present into item B.

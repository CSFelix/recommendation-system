<p style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>0.0 | Hypothesis</p>

So, our goal in this problem is to determine if the shounen animes average score and the seinen animes average score are equal. For this, we have two conclusions: the averages are equal and the averages are not equal. These two possible conclusions are called **Hypothesis**.

The first Hypothesis is called **Null Hypothesis** that always tell that nothing uncommon is happening, that is, it tells that the averages are equal and both data from shounen and seinen animes came from the same population.

In the other hand, we have the **Alternative Hypothesis** that always tell that something uncommon is happening, that is, it tells that the averages are not equal and both data from shounen and seinen animes came from different populations.

Statistically speaking, we have:

> **Null Hypothesis (H0)** - `anything uncommon is happening because the datas are from the same population`

$$
\text{H0}: \overline X(shounen) = \overline X(seinen)
$$

> **Alternative Hypothesis (H1)** - `something uncommon is happening because the datas are from different populations`

$$
\text{H1}: \overline X(shounen) < \overline X(seinen) \\
\lor \\
\text{H1}: \overline X(shounen) \neq \overline X(seinen) \\
\lor \\
\text{H1}: \overline X(shounen) > \overline X(seinen)
$$

<br />

Having this in mind, the possible outcomes we can achieve after applying the testings are:

> **Fail to Reject the Null Hypothesis (H0)** - `shounen and seinen animes have the same average score and are from the same population`;

> **Reject the Null Hypothesis** - `shounen and seinen animes have different average scores and are from different populations`.

<br />

Now, we are heading to where the Hypothesis Testing becomes interesting and I am going to blow up your mind. Even though we can fail to reject or reject the Null Hypothesis, there is the possibility that our result is wrong, right? For example, when we do a Math exam on school, even though we get a result in our calculations, there is the possibility that the result is wrong and we miss the question; the same thing can happen in Hypothesis Testing. We call them **Errors** and they are bummers ðŸ˜¥

There are two types of Errors and we will learn key concepts and techniques to decrease the possibility of making them. For now, let's see what their types are:

> **Error Type 1 - False Positive** - `the Null Hypothesis is rejected when it should be failed to be rejected. For instance, we concluded that the average score of shounen animes is different from the average score of seinen animes when they are actually equals`;

> **Error Type 2 - False Negative** - `the Null Hypothesis failed to be rejected when it should be rejected. For instance, we concluded that the average score of shounen animes and the average score of seinen animes are equal when they are actually differents`.

<p style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>0.1 | Level Regions</p>

Assume that we are talking to the MyAnimeList's owner and he asks us: *"So is the average score of shounen animes equal or different of the average score of seinen animes?"*, and we answer: *"With 100% of certainty and considering that the range of the average scores goes from 0 to 10, the averages are equals!"*. The owner will probably be pissed off on us. Why? Hehe...

Since the scores goes from 0 to 10, when we consider that the average range also goes from 0 to 10, all averages are equal, because their range will always be between 0 and 10, got you?

So, with a 100% of certainty, we can assume that the averages are the same, but since we are taking the minimum and the maximum possible values as the range boundaries, our conclusion is poor and contains no valuable insight. So, instead of having 100% of certainty and making a poor conclusion, what about we allow us to have some uncertainty about the result but make a richer conclusion? Are you going to take the risk? Yes? So do I!!

Normally, **95% (0.95)** is a common acceptable certainty whereas **5% (0.05)** is a common acceptable uncertainty. However, these percentages changes accordingly to the problem and to the risk the client and the Data Scientist decide to handle. For our plot, these percentages are fine!! So, now we are going to give a conclusion that has a chance of 95% (0.95) to be correct and 5% (0.05) to be wrong.

For our conclusion, we have to use the data to calculate the probability of getting a wrong conclusion if we choose to reject the Null Hypothesis when it should be failed to be rejected, that is, the probability of making a **Type 1 Error**. This probability is called **p-value**, goes from 0 to 1 and if it is smaller than the Significance Level, we reject the Null Hypothesis; otherwise, we fail to reject the Null Hypothesis.

In Statistics we call the acceptable certainty as **Confidence Level**, and the acceptable uncertainty as **Significance Level**, being:

> **Confidence Level** - `region/area on a Normal Distribution that shows the probability to the Null Hypothesis be failed to be rejected`;

> **Significance Level** - `region/area on a Normal Distribution that shows the probability to the Null Hypothesis be rejected`;

> **P-Value** - `probability of getting a wrong conclusion if we choose to reject the Null Hypothesis when it should be failed to be rejected. Statistically speaking, it is the probability of making a Type 1 Error`.

`When it is smaller than the Significance Level, it means that the probability of making a Type 1 Error is tiny and we can reject the Null Hypothesis. When it is equal or greater than the Significance Level, it means that the probability of making a Type 1 Error is large and we can fail to reject the Null Hypothesis.`

<p style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>0.2 | Calculating P-Values and Explaining P-Hacking</p>

Okay, you may be wondering: *"How to calculate the p-value in order to know if the Null Hypothesis must be rejected or fail to be rejected?"*. It depends on the type of Hypothesis Testing we are applying, but all types follow the same logic, they just add up:

1. the probability to observe the desired result in the Distribution - `it tells how rare is the desired result`;
2. the probability of obeservating other results that is equally rare to the desired result - `it tells how many rare events are likely the desired result`;
3. the probability of observating rarer or more extreme results to the desired result - `it tells how many rarer events are to the desired result`.

In a nutshell, the more observations of equally or rarer results, the higher is the p-value and the greater are the odds to fail to reject the Null Hypothesis.

---

Now, I am going to blow up your mind once again, even though the p-value is smaller enough to reject the Null Hypothesis, there are situations we cannot reject ðŸ¤¯. It happens due to an antiethical behavior called **P-Hacking**.

Picture this, consider that we get a sample of shounen and seinen animes, calculate their means, apply a T-Test (do not worry about it now, we will cover it later) for Hypothesis Testing and get a p-value higher than 0.05, then, we can say that the Null Hypothesis failed to be rejected.

But, just in order to get more confidence, we get another sample of shounen and seinen animes, do the same steps and conclude that we failed to reject H0. We do it over and over again and in all experiments get the same conclusion. After taking 429 experiments, we finally took one where the p-value is smaller than the Significance Level. So we concluded that we can finally reject H0, right? Nope! It is wrong!!

Even though in this last experiment we say we can reject H0, we did other 429 experiments comproving the inverse, so what do you think? It would be wrong and antiethical assuming only the result of the last experiment and discarding the result of the other experiments only to get the desired - and manipulated - Null Hypothesis rejection, right? The correct behavior here is to fail to reject the Null Hypothesis!!

I assume you already got the glimpse of how p-hacking works and why it is not cool to do. However, here is another example that you must avoid: we take another sample of shounen and seinen animes and got a p-value equals to 0.06. Since it is pretty close to the Significance Level, we add 2 or 3 more elements into our sample and get a p-value equals to 0.03, concluding that the Null Hypothesis must be rejected. Once again, it is anthietical and the correct behavior is to fail to reject H0.

Once an experiment is done, we cannot change the size and/or the elements of the sample only to have a chance to get the desired result, for instance, to have the chance to reject the Null Hypothesis when it should actually be failed to be rejected. To avoid it, there are techniques to calculate a great sample size for the testings and we will cover how to do it using **Statistical Power**.

> **P-Hacking: Number of Experiments** - `it is not ethical and allowed to take a bunch of experiments until to achieve the desired result. If the major experiments got a p-value higher than the Significance Level and only a few ones got a p-value smaller than, the correct behavior is to fail to reject H0`;

> **P-Hacking: Sample Size** - `also, it is not ethical and allowed to modify the sample size until to achieve the desired result`.

<p style='color:#7159c1; letter-spacing:2px; font-family:JetBrains Mono; font-size:150%'>0.3 | Statistical Power and Power Analysis</p>

Besides the **P-Hacking**, we have to deal with another problem: the situation where we failed to reject the Null Hypothesis when it should be rejected instead - **Error Type 2**. But do not worry, because **Statistical Power** and **Power Analysis** are here to solve both of them for us.

Kicking off with **Statistical Power**, it is the probability that we will correctly reject the Null Hypothesis, being **80% (0.80)** a common used value. Then, if we take a Hypothesis Testing and the choosen Statistical Power is 80%, we are literally saying that: *"If we conclude rejecting the Null Hypothesis, we have 80% of probability that our conclusion is correct."*. It may be kind of confusing when you think about it and the Confidence Level, but relax, I will wrap everything up in the end and all your doubts will disappear!!

So, just stick into your mind that Statistical Power is the probability of correct rejecting the Null Hypothesis and 80% is the standard value for many cases.

---

However, only determining that the Statistical Power is 80% nothing will change, we will still have to deal with p-hacking and the Error Type 2. So we use the **Power Analysis** in order to put the Statistical Power into action!!

With Power Analysis we determine a great size for our samples before taking the testings using the Statistical Power as a parameter. So at the same time we avoid p-hacking, we also avoid Error Type 2, killing two birds with one stone.

To calculate the Power Analysis for sample size, we have to follow four steps:

1. Determine the Power: `80% (0.80)`;

2. Determine the Significance Level: `5% (0.05)`;

3. Estimate the Overlap between the distributions;

4. Calculate the Sample Size accordingly to the Hypothesis Test to be Applied.

The overlap region between the two distributions must be quantified before calculating the sample size. When quantified, it is called **Effect Size (d)** and its equation is given as:

$$
\text{d} = \frac{ \left| \overline X(a) - \overline X(b) \right| } { \sqrt{\frac{\sigma(a)^2 + \sigma(b)^2} {2}} }
$$

where:

$$
\left| \overline X(a) - \overline X(b) \right| \text{: difference between the estimated means}
$$
